```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("SeoulBikeData.csv", encoding='unicode_escape')
df.head()
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], dayfirst=True)
df.head()
```

```python
df["day"] = df["date"].dt.day_name()
df["month"] = df["date"].dt.month
df["is_weekend"] = df["day"].isin(["Saturday", "Sunday"])
df.head()
```

```python
df.info()
df.describe()
df.isna().sum()
```

```python
df.describe()
```

```python
daily = df.groupby("date")["rented_bike_count"].sum().reset_index()
daily = daily.merge(df[["date", "seasons"]].drop_duplicates(), on="date")
plt.figure(figsize=(14,5))
sns.lineplot(data=daily, x="date", y="rented_bike_count", hue="seasons")
plt.title("Daily Bike Rentals Over Time (Colored by Season)")
plt.xlabel("Date")
plt.ylabel("Total Rentals Per Day")
plt.show()
```

Bike demand changes clearly with seasons.
Winter shows low rentals because of cold weather.
Summer has the highest rentals, showing people ride more in warm months.
The seasonal color coding shows the pattern more clearly than a single colored line.

```python
hourly = df.groupby(["date", "hour", "seasons"])["rented_bike_count"].mean().reset_index()
plt.figure(figsize=(14,6))
sns.lineplot(data=hourly, x="hour", y="rented_bike_count", hue="seasons")
plt.title("Hourly Average Bike Rentals (Colored by Season)")
plt.xlabel("Hour of the Day")
plt.ylabel("Average Rentals")
plt.show()
```

Most seasons show morning peak around 8–9 Am for full day rental.
Another evening peak appears around 5–7 Pm for whole night rental.
Summer line is highest because people prefer to ride bike in warm weather.
Winter line is very low because it is too cold for riding a bike.
Spring and Autumn show medium usage.

```python
temp_df = df.groupby(["temperature(°c)", "seasons"])["rented_bike_count"].mean().reset_index()
plt.figure(figsize=(14,6))
sns.lineplot(data=temp_df, x="temperature(°c)", y="rented_bike_count", hue="seasons")
plt.title("Temperature vs Average Bike Rentals (Colored by Season)")
plt.xlabel("Temperature(°C)")
plt.ylabel("Average Rentals")
plt.show()
```

Autumn shows the highest peak in bike rentals, even crossing 3000 rentals around 20–25°C. This means people ride the most during mild and comfortable autumn weather.
Summer also has high rentals, but the line is smoother and does not reach the extreme peak seen in autumn.
Spring shows moderate rentals that rise with temperature but remain below summer and autumn.
Winter consistently shows the lowest rentals across all temperature values due to cold and uncomfortable conditions.
Overall, rentals increase with temperature but drop again after very high temperatures (above ~30°C).


```python
humidity_df = df.groupby(["humidity(%)", "seasons"])["rented_bike_count"].mean().reset_index()
plt.figure(figsize=(14,6))
sns.lineplot(data=humidity_df, x="humidity(%)", y="rented_bike_count", hue="seasons", marker="o")
plt.title("Humidity vs Average Bike Rentals (Colored by Season)")
plt.xlabel("Humidity (%)")
plt.ylabel("Average Rentals")
plt.grid(True)
plt.show()
```

Bike rentals are highest when humidity is in the comfortable range of 20%–40%.
Very high humidity (above 80%) reduces rentals because the weather feels sticky and uncomfortable.
Winter shows the lowest rentals across all humidity levels.
Spring and summer have higher rental peaks because humidity is more comfortable during these seasons.
Overall, humidity significantly influences user comfort and bike rental activity.
