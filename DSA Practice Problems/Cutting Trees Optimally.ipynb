```python
import heapq
N = int(input())
H = list(map(int, input().split()))

if N < 3:
    print(0)
    exit()

left = [i - 1 for i in range(N)]
right = [i + 1 for i in range(N)]

left[0] = -1
right[N - 1] = -1

removed = [False] * N

def is_peak(i):
    if i < 0 or i >= N:
        return False
    if removed[i]:
        return False
    if left[i] == -1 or right[i] == -1:
        return False
    return H[i] > H[left[i]] and H[i] > H[right[i]]

heap = []
for i in range(N):
    if is_peak(i):
        heapq.heappush(heap, (-H[i], i))

total_wood = 0

while heap:
    neg_height, i = heapq.heappop(heap)
    height = -neg_height

    if not is_peak(i):
        continue

    total_wood += height
    removed[i] = True

    l = left[i]
    r = right[i]

    if l != -1:
        right[l] = r
    if r != -1:
        left[r] = l

    if l != -1 and is_peak(l):
        heapq.heappush(heap, (-H[l], l))

    if r != -1 and is_peak(r):
        heapq.heappush(heap, (-H[r], r))

print(total_wood)
```
