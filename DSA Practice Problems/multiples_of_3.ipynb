```python
N = int(input())
A = list(map(int, input().split()))
multiples = []
for num in A:
    if num % 3 == 0:
        multiples.append(num)

if len(multiples) == 0:
    print(-1)
else:
    print(*multiples)
```

```python
N = int(input())
A = list(map(int, input().split()))
current_count = 0
max_count = 0
for num in A:
    if num % 3 == 0:
        current_count += 1
        max_count = max(max_count, current_count)
    else:
        current_count = 0

print(max_count)
```
