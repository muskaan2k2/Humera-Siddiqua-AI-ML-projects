# 2. Perceptron

A perceptron is the simplest type of Artificial Neural Network. It is a basic building block of Deep Learning. It is used for binary classification problems. It gives only 2 outputs: (0 or 1), (yes or no), (True or false). A perceptron is like a small decision making machine. It: 1. Takes the input value, 2. Multiplies them with weights, 3. Adds them together, 4. Passes the result through an activation function, 5. Produces output(0 or 1). It works in a similar way how a neuron works in the human brain. 

### Mathematical Equation:

The main equation of a perceptron is: Z = w₁x₁ + w₂x₂ + w₃x₃ + ... + b (OR) Z = Σ (wᵢ xᵢ) + b. After that we apply activation function: y = f(Z), for step function: y = 0 if z ≥ 0 , y = 1 if z<0

### Explanation of each component

1. x₁, x₂, x₃ (Inputs): These are the features or values we give to the model. Ex: Hours studied, Attendence, Marks in assignments. They are the infromation given to the perceptron. 

2. w₁, w₂, w₃ (Weights): Weights tell us how important each input is, If weight is large -> input is very impoetant, If weight is small -> input is less important. The perceptron learns these weights during training.

3. b(bias): Bias is an extra value added to help the model shift the decision boundary. It helps the perceptron make better decisions even when inputs are zero. 

4. Z(weighted sum): z = (w₁x₁ + w₂x₂ + ... + b) This is the total score calculated before making final decision. 

5. Activation Function: In perceptron we use step function. It converts the value into: 0 or 1, If z is positive -> output = 1, If z is negative -> output = 0. This makes perceptron suitable for binary classification.

# 3. Activation Functions

### What is an Activation Function

An Activation function is a mathematical function used in Neural Networks. It decides weather a neuron should be activated or not. After calculating the weighted sum (z = wx+b), the activation function decides the final output. Without activation function: Neural Network behaves like a simple linear model, It cannot learn complex patterns. With activation function: The network can learn non-linear and complex relationships. 

### Purpose of activation function:

Activation Functions are used to: 

1. Introduce Non-linearity: Real world data is complex and non linear. Activation Functions help the network learn complex patterns. 

2. Decide Output Range: Some functions give output between: (0 to 1), (-1 to 1), (0 to infinity)

3. Help in Classification: For binary classification -> sigmoid, For deep networks -> ReLU, For Hidden layers -> Tanh or ReLU.                                                   Without activation Functions, a deep network would behave like a single layer linear model. 

### Sigmoid Function

##### Mathematical Formula:
σ(x) = 1 / (1 + e^(-x))

##### Output Range:
0 to 1

##### Shape:
S-shaped curve

##### Where it is used:
- Output layer for binary classification
- Probability prediction

Example:
If output = 0.8 → 80% probability


##### Advantages of Sigmoid

1. Output between 0 and 1
2. Good for probability interpretation
3. Smooth gradient

##### Disadvantages of Sigmoid

1. Vanishing Gradient Problem
   For large positive or negative values,
   gradient becomes very small.

2. Output is not zero-centered
   Makes learning slower.

3. Slow training for deep networks


### ReLU Function

##### Mathematical Formula:

f(x) = max(0, x)

##### Output Range:
0 to infinity

##### Shape:
Flat for negative values
Linear for positive values

##### Where it is used:
- Hidden layers in Deep Learning
- CNNs

##### Advantages of ReLU

1. Very simple and fast
2. Reduces Vanishing Gradient Problem
3. Faster training
4. Works very well in deep networks

##### Disadvantages of ReLU

1. Dying ReLU Problem
   If neuron gets negative values always,
   it may stop learning.

2. Output not bounded
   Can give very large values


### Tanh Function

##### Mathematical Formula:

tanh(x) = (e^x - e^(-x)) / (e^x + e^(-x))

##### Output Range:
-1 to +1

##### Shape:
S-shaped curve (like sigmoid but centered at zero)

##### Where it is used:
- Hidden layers
- When negative values are needed

##### Advantages of Tanh

1. Zero-centered output (-1 to 1)
2. Stronger gradients than sigmoid
3. Works better in hidden layers

##### Disadvantages of Tanh

1. Still suffers from Vanishing Gradient Problem
2. Slower than ReLU


# 4. Architecture of Basic Artificial Neural Network(ANN)

An Artificial Neural Network (ANN) is a machine learning model inspired by the human brain.
Just like our brain has neurons, ANN has artificial neurons.

These neurons:
- Receive input
- Process the information
- Produce output

ANN is mainly used for:
- Classification
- Regression
- Image recognition
- Speech recognition

It learns patterns from data using training.

### Basic Architecture of ANN

A basic ANN has three main layers:

1. Input Layer
2. Hidden Layer
3. Output Layer

Information flows from:
Input → Hidden → Output

This flow is called Forward Propagation.

### Input Layer 
The Input Layer is the first layer of the neural network.

It receives the data.

Each neuron in the input layer represents one feature.

Example:

If we want to predict house price:
- Size of house
- Number of rooms
- Location rating

Then we have 3 input neurons.        

### Hidden Layer 
The Hidden Layer is between input and output layer.

It performs calculations.

Each neuron in hidden layer:

1. Takes input values
2. Multiplies by weights
3. Adds bias
4. Applies activation function

There can be:
- One hidden layer (Shallow Network)
- Multiple hidden layers (Deep Network)

Hidden layers help the network learn complex patterns.

### Output Layer
The Output Layer gives the final result.

It depends on the problem type:

Binary Classification:
1 neuron (0 or 1)

Multi-class Classification:
Multiple neurons (one for each class)

Regression:
1 neuron (continuous value)
Example:
Spam detection → Output = 0 or 1

### Weights
Weights tell how important each input is.

If weight is large:
Input has more importance.

If weight is small:
Input has less importance.

The model learns these weights during training.

### Bias
Bias is an extra value added to the weighted sum.

It helps shift the decision boundary.

Formula:

z = (w₁x₁ + w₂x₂ + ... + wₙxₙ) + b

### Forward Propagation 
Forward propagation means:Passing input data through the network to get the output.

Steps:

Step 1:
Multiply inputs with weights

Step 2:
Add bias

Step 3:
Apply activation function

Step 4:
Pass result to next layer

Step 5:
Repeat until output layer

Mathematical Representation:

Hidden layer:
z₁ = W₁X + b₁
a₁ = activation(z₁)

Output layer:
z₂ = W₂a₁ + b₂
y = activation(z₂)

This process is called Forward Propagation.

# 5 Backpropagation:

Backpropagation is a training algorithm used in Artificial Neural Networks.

Backpropagation is the method used to teach the neural network how to correct its mistakes.

When a neural network makes a prediction:
- Sometimes it is correct.
- Sometimes it is wrong.

Backpropagation helps the network:
1. Measure how wrong it is.
2. Find out which weights caused the error.
3. Adjust those weights.
4. Improve the prediction next time.

In simple words:

Forward propagation → makes prediction  
Backpropagation → corrects the mistake  

This process repeats many times until the model becomes accurate.

### Role of chain rule in updating weights:

To update weights, we need to calculate gradients.

Gradient means:
How much the output changes when weight changes.

But in neural networks:
Output depends on hidden layer.
Hidden layer depends on weights.

So we have multiple layers connected.

Here comes the Chain Rule.


### What is Chain Rule?

Chain Rule is a rule from calculus.

It helps us calculate the derivative of functions that are connected together.

Simple example:

If:
A affects B
B affects C

Then to know how A affects C,
we multiply:

(dC/dB) × (dB/dA)

This is Chain Rule.

### Why Gradient Descent is used?
Gradient Descent is used to minimize loss by updating weights
in the opposite direction of the gradient.

Without backpropagation and gradient descent,
neural networks cannot learn. 
1. It reduces error step by step.
2. It helps model reach minimum loss.
3. It makes learning automatic.
4. Without it, weights will not improve.

If learning rate is:
Too small → learning is very slow  
Too large → model may not converge  

# 6. Batch Gradient Descent vs Stochastic Gradient Descent(SGD) vs Mini-Batch Gradient Descent

### Batch Gradient Descent

Batch Gradient Descent uses the entire training dataset to calculate the gradient before updating the weights.
##### How it works:
Step 1:
Take all training examples.

Step 2:
Calculate total loss using entire dataset.

Step 3:
Compute gradient.

Step 4:
Update weights once.

This process repeats for many epochs.
##### Example:
If dataset has 10,000 samples:
Batch Gradient Descent will use all 10,000 samples
before making one weight update.
##### Advantages:
1. Very stable updates.
2. Smooth convergence.
3. Accurate gradient calculation.
##### Disadvantages:
1. Very slow for large datasets.
2. Requires high memory.
3. Not practical for very big data.

### Stochastic Gradient Descent (SGD)

Stochastic Gradient Descent updates weights 
using only one training example at a time.

"Stochastic" means random.

##### How it works:

Step 1:
Take one training sample.

Step 2:
Calculate loss.

Step 3:
Compute gradient.

Step 4:
Update weights immediately.

Then move to next sample.

##### Example:

If dataset has 10,000 samples:

SGD will update weights 10,000 times in one epoch.

##### Advantages:

1. Very fast updates.
2. Works well for large datasets.
3. Requires less memory.
4. Can escape local minima.

##### Disadvantages:

1. Updates are noisy.
2. Loss graph is not smooth.
3. May take longer to converge properly.

### Mini-batch Gradient Descent

Mini-batch Gradient Descent is a combination of 
Batch and Stochastic Gradient Descent.

Instead of using:
- Entire dataset (Batch)
- One sample (SGD)

It uses a small group of samples called a mini-batch.


##### How it works:

Step 1:
Divide dataset into small batches (e.g., 32, 64, 128 samples).

Step 2:
Calculate loss for one mini-batch.

Step 3:
Compute gradient.

Step 4:
Update weights.

Repeat for all mini-batches.

##### Example:

If dataset has 10,000 samples
and batch size is 100:

Weights are updated 100 times in one epoch.

##### Advantages:

1. Faster than Batch Gradient Descent.
2. More stable than SGD.
3. Efficient for large datasets.
4. Works very well with GPUs.

##### Disadvantages:

1. Slightly noisy updates.
2. Need to choose proper batch size.

# 7 Overfitting in Neural Network

### Overfitting
Overfitting is a common problem in neural networks and machine learning models.

It happens when a model learns the training data too well, including the noise and small details that are not important.

Because of this, the model performs:

- Very well on training data
- Poorly on new or unseen data

The model memorizes the training data instead of learning general patterns.

### Techniques to Prevent Overfitting

##### Technique 1: Dropout

Dropout is a regularization technique used to reduce overfitting.

During training, dropout randomly turns off some neurons in the network.
This means some neurons are temporarily ignored while training.

##### Why this helps

Because the network cannot depend on specific neurons, it learns more general and robust patterns.

##### Example

If a layer has 100 neurons and dropout rate is 0.5,
then 50 neurons are randomly ignored during each training step.
This prevents the network from memorizing the training data.

### Technique 2: Early Stopping

Early stopping means stopping the training process before the model starts overfitting.

During training we monitor:

- Training loss
- Validation loss

Normally:

Training loss keeps decreasing.

But after some time, validation loss starts increasing.

This means the model is starting to overfit.

So we stop the training at that point.

##### Benefit

The model keeps the best weights before overfitting begins.

### Technique 3: Data Augmentation

Data augmentation is used mainly in image datasets.

It means creating more training data by modifying existing images.

For example:

- Rotating images
- Flipping images
- Zooming
- Cropping
- Changing brightness

This increases the size and diversity of the dataset.

##### Why this helps

More data helps the model learn general patterns instead of memorizing images.

# 8 Convolutional Neural Network

A Convolutional Neural Network (CNN) is a type of deep learning model that is mainly used to process image data.

CNNs are designed to automatically detect important features from images, such as edges, shapes, textures, and objects.

Unlike traditional neural networks, CNNs do not treat images as simple numbers.  
Instead, they preserve the spatial structure of the image (the arrangement of pixels).

A CNN uses special operations called convolutions that scan the image using small filters to detect patterns.

Because of this ability, CNNs are widely used in tasks like:

- Image classification  
- Face recognition  
- Object detection  
- Medical image analysis  
- Self-driving car vision systems

In simple words, a CNN works like a visual recognition system, similar to how humans recognize objects in images.

### Why CNNs Are Preferred for Image Data

CNNs are preferred for image data because they are specially designed to handle the structure of images.

##### 1. Automatic Feature Extraction

In traditional machine learning, features must be extracted manually.

But CNNs automatically learn important features from images during training.

This reduces human effort and improves accuracy.

##### 2. Preserves Spatial Relationships

Images contain spatial information (neighboring pixels are related).

CNNs preserve this spatial structure while processing images.

This helps the model understand shapes and patterns better.

##### 3. Parameter Sharing

CNNs use the same filter across the entire image.

This means fewer parameters compared to fully connected networks.

Because of this, CNNs are more efficient and easier to train.

##### 4. Translation Invariance

CNNs can recognize objects even if they appear in different positions in the image.

For example, a CNN can still recognize a cat whether it appears at the top, bottom, or center of the image.

# 9 Components of Convolutional Neural Network (CNNs)

Convolutional Neural Networks (CNNs) are made up of several important components that help the network understand and process image data.

Each component has a specific role in extracting useful information from images.

The main components of CNNs include:

- Convolution Layer
- Filter (Kernel)
- Stride
- Padding
- Pooling Layer

These components work together to detect patterns, reduce image size, and improve learning efficiency.

### (a) Convolution Layer

The Convolution Layer is the most important layer in a CNN.

Its main purpose is to extract features from the input image.

Instead of looking at the whole image at once, this layer scans the image using small filters.

These filters move across the image and detect patterns such as:

- edges
- lines
- textures
- shapes

For example:

In early layers, the convolution layer may detect simple features like edges.

In deeper layers, it may detect complex features like eyes, faces, or objects.

Because of this process, the model can understand important visual patterns in images.

### (b) Filter / Kernel

A Filter (also called a Kernel) is a small matrix of numbers used in the convolution process.

It slides across the image and performs element-wise multiplication with the image pixels.

The result of this operation produces a feature map, which highlights important features in the image.

For example:

A filter may be designed to detect:

- vertical edges
- horizontal edges
- textures

During training, the neural network automatically learns the best filter values that help detect useful patterns.

Different filters can detect different types of features from the same image.

### (c) Stride

Stride refers to the number of pixels the filter moves across the image at each step.

For example:

Stride = 1  
The filter moves one pixel at a time.

Stride = 2  
The filter moves two pixels at a time.

Larger stride values reduce the size of the output feature map.

This helps decrease computation and makes the network faster.

However, if the stride is too large, the model might miss some important image details.

### (d) Padding

Padding means adding extra pixels (usually zeros) around the border of the image.

This is done before applying the convolution operation.

Padding helps in two important ways:

1. It prevents the image from shrinking too quickly after convolution.
2. It allows filters to process the edge pixels of the image properly.

Without padding, important information near the edges of the image may be lost.

Common types of padding include:

- Valid Padding (no extra pixels added)
- Same Padding (output size remains the same as input size)

### (e) Pooling Layer

The Pooling Layer is used to reduce the size of the feature maps.

It helps make the network more efficient by reducing the number of parameters and computations.

Pooling keeps the most important information while removing unnecessary details.

The most common type is Max Pooling.

In Max Pooling:
The maximum value from a small region of the feature map is selected.

For example:

From a 2×2 region, the largest value is chosen.

Benefits of pooling include:

- Reducing computational cost
- Preventing overfitting
- Making the model more robust to small changes in the image

# 10 Output Dimention of a Convolution Layer

##### Given
- Input Image Size = 32 × 32 × 3
- Number of Filters = 10
- Filter (Kernel) Size = 3 × 3
- Stride (S) = 1
- Padding (P) = 0 (No padding)


##### Formula Used

The output size of a convolution layer is calculated using the formula:

[Output\ Size = \frac{(N - F + 2P)}{S} + 1]

Where:

- **N** = Input size (height or width)
- **F** = Filter size
- **P** = Padding
- **S** = Stride

##### Step 1: Calculate Output Height and Width

Substitute the values into the formula:

[Output = \frac{(32 - 3 + 2(0))}{1} + 1]

[= \frac{29}{1} + 1]

[= 30]

So,

- **Output Height = 30**
- **Output Width = 30**

##### Step 2: Calculate Output Depth

The depth of the output is equal to the number of filters used in the convolution layer.

Number of Filters = **10**

So,

- **Output Depth = 10**

### Final Output Dimension

[mathbf{30 \times 30 \times 10}]
