39) Decision boundary in logistic regression.
In Logistic regression, Decision Boundary is a line that separates 2 classes.
it is used when output is Yes / no.
Ex: on one side - class 0(No), on other side - class 1(yes).
This line decides which side is class 0 , and which side is class 1.

Logistic Regression gives probablity between 0 & 1, If probablity is greater than or equal to 0.5 - it is class 1,  If probablity is less than 0.5 - it is class 0. The decision boundary is where probablity = 0.5

import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

Creating dataset

```python
X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([0, 0, 0, 1, 1, 1])
```

Dataset Explanation: students studying 1-3 hours - fails, students studying 4-6 hours - pass

Training Model

```python
model = LogisticRegression()
model.fit(X, y)
```

Model learns at which point fail changes to pass

Get Probablities

```python
X_test = np.linspace(0, 7, 100).reshape(-1, 1)
probabilities = model.predict_proba(X_test)[:, 1]
```

We check probabilities for many values from 0 to 7 hours.

Decision Boundary

```python
decision_boundary = -model.intercept_ / model.coef_
decision_boundary
```

this gives the exact point where probability=0.5, That point is the decision boundary.

Plotting graph to see everything clearly

```python
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_test, probabilities, color='green', label='Probability curve')
plt.axvline(x=decision_boundary, color='red', linestyle='--', label='Decision curve')
plt.xlabel('Hours Studied')
plt.ylabel('Probability of passing')
plt.legend()
plt.show()
```

here red line is decision boundary, left of red line is fail(class 0), right of red line is pass (class 1)
