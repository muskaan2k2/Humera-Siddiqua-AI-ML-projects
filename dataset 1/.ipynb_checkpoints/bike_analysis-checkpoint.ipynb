```python
sns.pairplot(df)
plt.show()
```

```python
cols = ['Rented Bike Count','Temperature(Â°C)','Humidity(%)','Wind speed (m/s)','Visibility (10m)','Solar Radiation (MJ/m2)','Rainfall(mm)','Snowfall (cm)']
sns.pairplot(df[cols])
plt.show()
```

```python
df.hist(figsize=(15,10), bins=30)
plt.show()
```

```python
plt.scatter(df['Temperature(Â°C)'], df['Rented Bike Count'])
plt.xlabel("Temperature")
plt.ylabel("Bike Count")
plt.title("Temperature vs Bike Rentals")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['Date'], df['Rented Bike Count'])
plt.title("Bike Rentals Over Time")
plt.show()
```

```python
sns.barplot(x='Hour', y='Rented Bike Count', data=df)
plt.show()
```
