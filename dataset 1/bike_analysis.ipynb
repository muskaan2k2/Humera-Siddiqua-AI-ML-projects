```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("SeoulBikeData.csv", encoding='unicode_escape')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], dayfirst=True)
df.head()
```

```python
df.columns
```

```python
sns.pairplot(df)
plt.show()
```

The pairplot shows how every variable in the dataset is related to every other variable. 
We can clearly see that temperature, dew point temperature, and solar radiation have a positive relationship with bike rentals, meaning rentals increase when the weather becomes nicer.
Humidity, wind speed, rainfall, and snowfall mostly show a negative or unclear pattern with rentals, because bad weather reduces people's willingness to ride bikes.
This plot helps us quickly see which weather factors affect rentals the most and which ones do not have a strong influence.

```python
plt.scatter(df['temperature(°c)'], df['rented_bike_count'])
plt.xlabel("Temperature")
plt.ylabel("Bike Count")
plt.title("Temperature vs Bike Rentals")
plt.show()
```

The scatter plot shows that bike rentals increase as temperature increases. When the temperature is cold (below 0°C), the number of rentals is very low. As the temperature becomes pleasant (10°C to 25°C), rentals increase sharply. The highest rentals are seen around 20°C to 30°C because this is the most comfortable temperature for riding bike. After around 30°C, rentals start dropping a little because it becomes too hot. This clearly shows that temperature has a strong and direct effect on bike rental activity.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x='temperature(°c)',
    y='rented_bike_count',
    hue='seasons'
)
plt.title("Temperature vs Bike Rentals (Colored by Seasons)")
plt.xlabel("Temperature (°C)")
plt.ylabel("Rented Bike Count")
plt.show()
```

When the weather is warm (15°C–30°C), people rent more bikes. When the weather is Cold (0°C to –15°C), people rent very few bikes. Dots for summer and autumn are highest on the graph. People avoid biking in extreme heat, and cold. This scatter plot uses three variables: temperature, bike count, and season. 
Each season has a different color, so we can easily see how different seasons affect bike rentals. 
The first scatter plot uses only two variables: temperature and bike count. 
This makes the plot simpler but less informative which does not use color for seasons, so all points look the same. 
While this plot helps us compare trends between seasons.


```python
df.hist(figsize=(15,10), bins=30)
plt.show()
```

This graph shows how often each value appears in the dataset. Most hours have no rain or snow. 
Temperature is usually in a normal comfortable range. Wind speed and humidity mostly stay in the 
middle values. Rented bike count is usually low but sometimes becomes very high on good-weather days. 
Overall, this helps us understand what values are common, what values are rare, and how the data is spread.

```python
plt.figure(figsize=(10,6))
sns.kdeplot(
    data=df,
    x="temperature(°c)",
    hue="seasons",
    fill=True,
)
plt.title("Temperature Distribution by Season")
plt.xlabel("Temperature (°C)")
plt.ylabel("Density")
plt.show()
```

This KDE plot shows how temperature changes in each season. If the curve is tall the temperature happens many times.
If the curve is short the temperature happens rarely.
Winter has mostly very low temperatures, while summer has high temperatures. 
Spring and autumn stay in the middle range. KDE is better than a histogram because it gives a clean, smooth shape and makes comparison between seasons easier. Histograms can look messy when many colors overlap, but KDE curves 
stay clear and easy to understand.

```python
sns.barplot(x='hour', y='rented_bike_count', data=df)
plt.show()
```

The bar plot shows how bike rentals change during the 24 hours of the day. 
The number of rentals is very low during the night (from 0:00 to 5:00 AM) because most 
people are sleeping and very few are travelling at that time.
There is a sharp increase around 7:00–9:00 AM, which matches the morning office and school 
rush hours. This shows that many people use bikes for commuting in the morning.
During midday (11:00 AM to 3:00 PM), the rental numbers stay moderate because people travel 
less compared to peak office hours.
In the evening, there is another big peak around 5:00–7:00 PM, which matches the time people 
return home from work or school. This confirms that bikes are used heavily during commuting times.
After 8:00 PM, rentals drop again as the day ends. Overall, the bar plot clearly shows two main 
peak periods morning and evening rush hours proving that bike rentals are strongly connected to daily travel routines.

```python
plt.figure(figsize=(12,6))
sns.barplot(
    x='hour',
    y='rented_bike_count',
    hue='seasons',
    data=df
)
plt.title("Hourly Bike Rentals by Season")
plt.xlabel("Hour of the Day")
plt.ylabel("Average Bike Rentals")
plt.show()
```

This 3-variable bar plot compares bike rentals across 24 hours for each season. 
Summer and autumn show the highest rental activity throughout the day, especially 
around 7–9 AM and 5–7 PM, which matches daily commute times. Winter has the lowest 
rentals regardless of the hour because of cold weather. Spring stays in between. 
Night hours (0–5 AM) show very few rentals in all seasons. This plot clearly shows 
how both hour and season together influence bike usage. The first bar plot uses only two variables (Hour and Bike Count). It shows the overall hourly pattern but does not show how seasons change the rentals.
The three-variable bar plot adds 'season' as an extra variable using different colors. This helps us compare hourly bike rentals for each season separately. It shows that summer and autumn have the highest rentals, while winter has the lowest. This graph provides deeper understanding and clearer seasonal differences compared to the first bar plot.

```python
plt.figure(figsize=(12,5))
plt.plot(df['seasons'], df['rented_bike_count'])
plt.title("Bike Rentals Over Time")
plt.show()
```

The line plot shows how bike rentals change over the entire time period in the dataset. 
We can see that rentals are very low during winter months because the weather is cold. 
As the seasons change to spring and summer, rentals increase sharply. 
The highest rentals are found in warmer months, and then the usage drops again during autumn 
as the temperature begins to fall.
This graph clearly shows a strong seasonal pattern — rentals rise and fall depending on the 
time of year. It helps understand long-term trends and how real-life seasons affect customer behavior.

```python
daily = df.groupby("date")["rented_bike_count"].sum().reset_index()
daily = daily.merge(df[["date", "seasons"]].drop_duplicates(), on="date")
plt.figure(figsize=(14,5))
sns.lineplot(data=daily, x="date", y="rented_bike_count", hue="seasons")
plt.title("Daily Bike Rentals Over Time (Colored by Season)")
plt.xlabel("Date")
plt.ylabel("Total Rentals Per Day")
plt.show()
```

This line plot shows how the total number of bikes rented changes every day, and it also shows which season each day belongs to. We can clearly see that summer and autumn days have high rental numbers, while winter days have low rentals. The colors help us understand seasonal changes: 
warm seasons create higher peaks, and cold seasons create lower lines. This graph shows the daily trend more clearly by combining both date and season information.

```python
hourly = df.groupby(["date", "hour", "seasons"])["rented_bike_count"].mean().reset_index()
plt.figure(figsize=(14,6))
sns.lineplot(data=hourly, x="hour", y="rented_bike_count", hue="seasons")
plt.title("Hourly Average Bike Rentals (Colored by Season)")
plt.xlabel("Hour of the Day")
plt.ylabel("Average Rentals")
plt.show()
```

Most seasons show morning peak around 8–9 Am for full day rental. Another evening peak appears around 5–7 Pm for whole night rental. Summer line is highest because people prefer to ride bike in warm weather. Winter line is very low because it is too cold for riding a bike. Spring and Autumn show medium usage. We use this graph to understand how bike rentals change during the day and how seasons affect hourly rental patterns. It helps us see when people ride bikes the most, and which season has the highest or lowest hourly usage. This is useful for planning bike availability,managing 
rush hours, and understanding user behavior across seasons.

```python
temp_df = df.groupby(["temperature(°c)", "seasons"])["rented_bike_count"].mean().reset_index()
plt.figure(figsize=(14,6))
sns.lineplot(data=temp_df, x="temperature(°c)", y="rented_bike_count", hue="seasons")
plt.title("Temperature vs Average Bike Rentals (Colored by Season)")
plt.xlabel("Temperature(°C)")
plt.ylabel("Average Rentals")
plt.show()
```

Autumn shows the highest peak in bike rentals, even crossing 3000 rentals around 20–25°C. This means people ride the most during mild and comfortable autumn weather.
Summer also has high rentals, but the line is smoother and does not reach the extreme peak seen in autumn.
Spring shows moderate rentals that rise with temperature but remain below summer and autumn.
Winter consistently shows the lowest rentals across all temperature values due to cold and uncomfortable conditions.
Overall, rentals increase with temperature but drop again after very high temperatures (above ~30°C).

```python
humidity_df = df.groupby(["humidity(%)", "seasons"])["rented_bike_count"].mean().reset_index()
plt.figure(figsize=(14,6))
sns.lineplot(data=humidity_df, x="humidity(%)", y="rented_bike_count", hue="seasons", marker="o")
plt.title("Humidity vs Average Bike Rentals (Colored by Season)")
plt.xlabel("Humidity (%)")
plt.ylabel("Average Rentals")
plt.grid(True)
plt.show()
```

Bike rentals are highest when humidity is in the comfortable range of 20%–40%.
Very high humidity (above 80%) reduces rentals because the weather feels sticky and uncomfortable.
Winter shows the lowest rentals across all humidity levels.
Spring and summer have higher rental peaks because humidity is more comfortable during these seasons.
Overall, humidity significantly influences user comfort and bike rental activity.
