```python
import heapq
```

Creating Two empty heaps

```python
max_heap = []
min_heap = []
```

Function to add number into heap

```python
def add_num(num):
    if not max_heap or num <= -max_heap[0]:
        heapq.heappush(max_heap, -num)
    else:
        heapq.heappush(min_heap, num)
```

balancing the heaps

```python
def balance():
    if len(max_heap) > len(min_heap) + 1:
        heapq.heappush(min_heap, -heapq.heappop(max_heap))
    elif len(min_heap) > len(max_heap):
        heapq.heappush(max_heap, -heapq.heappop(min_heap))
```

get median

```python
def get_median():
    if k % 2 == 1:
        return -max_heap[0]
    else:
        return (-max_heap[0] + min_heap[0]) / 2
```

```python
arr = [10, 20, 30, 40, 100]
k = 3
```

defining input array and window size

```python
result = []

for i in range(len(arr)):
    add_num(arr[i])
    balance()

    if i >= k:
        median = get_median()
        result.append(abs(arr[i] - median))
```

2. top-k correlated feature pairs

import required libraries

```python
import math
import heapq
```

input data

```python
data = [
    [1, 2, 3],
    [2, 4, 6],
    [3, 6, 9],
    [4, 8, 12]
]
k = 1
```

convert rows to columns

```python
num_rows = len(data)
num_features = len(data[0])

features = list(zip(*data))
```

precompute mean of each feature

```python
means = []

for col in features:
    mean = sum(col) / num_rows
    means.append(mean)
```

precompute SD of each feature

```python
stds = []

for i in range(num_features):
    variance = sum((x - means[i]) ** 2 for x in features[i]) / num_rows
    stds.append(math.sqrt(variance))
```

Compute correlation between feature pairs

```python
min_heap = []

for i in range(num_features):
    for j in range(i + 1, num_features):
        
        covariance = sum(
            (features[i][r] - means[i]) * (features[j][r] - means[j])
            for r in range(num_rows)
        ) / num_rows
        
        if stds[i] == 0 or stds[j] == 0:
            continue
        
        corr = covariance / (stds[i] * stds[j])
        abs_corr = abs(corr)
        
        if len(min_heap) < k:
            heapq.heappush(min_heap, (abs_corr, i, j, corr))
        else:
            if abs_corr > min_heap[0][0]:
                heapq.heappop(min_heap)
                heapq.heappush(min_heap, (abs_corr, i, j, corr))
```

Extract and Format Output

```python
result = []

while min_heap:
    _, i, j, corr = heapq.heappop(min_heap)
    result.append((i, j, round(corr, 3)))

result.reverse()

result
```
